```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
provider=pd.read_csv('..\data\Medicare_Provider_Util_Payment_PUF_CY2017.txt', sep='\t', header = 0,
                      usecols=['npi','nppes_provider_last_org_name',
             'nppes_provider_first_name','nppes_provider_mi','nppes_credentials' ,
             'nppes_entity_code','place_of_service','provider_type',
             'nppes_provider_zip','nppes_provider_city','nppes_provider_state',
             'hcpcs_code','hcpcs_description','bene_unique_cnt','bene_day_srvc_cnt',                 
             'average_Medicare_allowed_amt','average_Medicare_payment_amt'],
    dtype ={'npi': 'str','nppes_provider_zip': 'str', 
            'nppes_entity_code':'category',
           'place_of_service': 'category',
           'provider_type':'category'},skiprows=[1])

provider.head()
```

```python
provider = provider.rename(columns={'nppes_provider_last_org_name':'np_last_orgname',
             'nppes_provider_first_name':'np_firstname',
             'nppes_provider_mi':'np_midname',
             'nppes_credentials':'np_credentials' ,
             'nppes_entity_code':'np_entity_code',
             'provider_type':'p_type',
             'nppes_provider_zip': 'np_zip',
             'nppes_provider_city': 'np_city',
             'nppes_provider_state': 'np_state',
             'hcpcs_code':'HCPCS',
             'bene_unique_cnt':'unique_beneficiary',                          
             'average_Medicare_allowed_amt':'avg_medicare_allowed',
             'average_Medicare_payment_amt':'avg_medicare_pymt'
})
                             
```

```python
#providers with entity type 'O' and place of service 'F'
provider_Org_F = provider[(provider.np_entity_code=='O') & (provider.place_of_service == 'F')]
provider_Org_F =provider_Org_F.drop(['np_firstname', 'np_midname', 'np_credentials'], axis=1)
```

###### There're 57329 records with provider type("Ambulatory Surgical Center) for a combination of entity type 'O' and place of service 'F'; There's no record for this provider type in provider with entity type 'O' and place of service 'O'

```python
# only this combination of entity type "O" and place of service "F" has provider type "Ambulatory Surgical Center"
provider_ASC = provider_Org_F[provider_Org_F.p_type=="Ambulatory Surgical Center"]
provider_ASC.to_csv('provider_ASC.csv', index=False)
```
